```python
import pandas as pd 
import os 
import numpy as np
import json
```

## Overall

```python
list_files = [file_ for file_ in os.listdir("DUMP") if file_.endswith("RESULTS.json")]
json_files = []
for json_file in list_files:
    print(json_file)
    with open("DUMP/" + json_file, "r") as f:
        js = json.load(f)
        json_files.append(js)
```

```python
json_files[0][0].keys()
```

```python
keys_to_exclude = ["predictions", "labels", "generations", "unfiltered_outputs"]
```

```python
type(json_files[0][0])
```

```python
for time in json_files:
    for d in time:
        for key in keys_to_exclude:
            d.pop(key, None)
```

```python
json_files[0][0].keys()
```

```python
json_files = [item for sublist in json_files for item in sublist]
```

```python
len(json_files)
```

```python
df = pd.DataFrame(data=json_files, columns=json_files[0].keys())
```
